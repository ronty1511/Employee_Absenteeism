```python
data_targets = data_preprocessed.drop(['Absenteeism Time in Hours'], axis=1)
```

```python
data_targets
```

```python
raw_features = data_targets.iloc[:, 0:-1]
raw_features
```

```python
#It is vital to standardize the dataset for it to be used in a machine learning model. SKLearn has builtin functions that 
#serve the purpose. Standardization is subtracting the mean from every value and dividing the result by the standard
#deviation. Can be interpreted as centering and scaling the data. However since the dataset consists of dummy variables
#too, we have to design a custom scaler that doesn't standardize the dummy variables.

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler=StandardScaler(copy, with_mean, with_std)
        self.columns=columns
        self.mean_=None
        self.var_=None
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_=np.mean(X[self.columns])
        self.var=np.var(X[self.columns])
        return self
    def transform(self, X, y=None, copy=None):
        init_col_order=X.columns
        std_X=pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        raw_X=X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([raw_X, std_X], axis=1)[init_col_order]
```

```python
raw_features.columns.values
```

```python
columns_to_be_standardized=['Month', 'Day', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Children', 'Pets']
```

```python
absenteeism_scaler=CustomScaler(columns_to_be_standardized)
```

```python
#the .fit function feeds the StandardScaler with our data. The means and std. devns. of each column are stored in the
#StandardScaler object which can be used later.
absenteeism_scaler.fit(raw_features)
```

```python
std_features = absenteeism_scaler.transform(raw_features)
```

```python
std_features.shape
```

```python
#It is about time to split the dataset into train set and test set. Again sklearn comes to make things easier for us.
#train_test_split from sklearn can split the data into 4 subgroups (X_train, X_test, y_train, y_test). We can also4
#specify the ratio we want to split it in.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(std_features, targets, train_size=0.8)

#setting random_state to a number will get rid of shuffling.
```

```python
#Now that standardizing and splitting is over we can proceed with fitting the data in the model

from sklearn.linear_model import LogisticRegression
from sklearn import metrics

reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
#Accuracy of the model is 75.71%, i.e., it has been able to detect 75% of the data correctly
reg.score(X_train, y_train)
```

```python
#Let us dig deeper into the model and find out the coefficients and intercepts of the logistic
#regression module.

print(reg.intercept_, reg.coef_, sep='\n\n') #Returns the intercept and coefficients respectively
```

```python
#Let us create a summary table now where each columns coefficient is represented in tabulated form along with
#the log odds. log(h)=b0+b1x1+b2x2+...

feature_names = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month', 'Day',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets']
summary_table = pd.DataFrame(columns=['Feature Name'], data=feature_names)
summary_table['Coefficient']=reg.coef_.T
summary_table.index=summary_table.index+1
summary_table.loc[0]=['Intercept', reg.intercept_[0]]
summary_table['Log h']=np.exp(summary_table['Coefficient'])
summary_table
```

```python
#Sorting the table with the highest value of log odds at the top
summary_table=summary_table.sort_values('Log h', ascending=False)
summary_table
#Coefficients that are far away from zero (on both sides) are considered more important which, in this case,
#are the ones at the top and the ones at the bottom.
```
