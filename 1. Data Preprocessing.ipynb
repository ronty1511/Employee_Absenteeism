```python
#It can be inferred that 'Reason for Absence' column can contain values from 0 to 28 (excluding 20).
```

```python
#A dummy variable is equal to 1 if entry falls into a certain category and 0 otherwise. It is used in explanatory
#data analysis.
```

```python
reasons_col = pd.get_dummies(df['Reason for Absence'])
reasons_col
```

```python
#0 in Reasons for Absence implies that a particular entry does not have any disease. Since this feature won't help us
#predict the hours of absenteeism, it should be dropped.
reasons_col = reasons_col.drop([0], axis=1)
reasons_col
```

```python
#The next step is to group the reasons into categories and add these categories to the main dataframe
```

```python
#type1 : Various diseases
#type2 : Pregnancy and childbirth
#type3 : Poisoning and uncategorized
#type4 : Less serious reasons, such as a doctor appointment, physiotherapy

type1 = reasons_col.loc[:, 1:14].max(axis=1)
type2 = reasons_col.loc[:, 15:17].max(axis=1)
type3 = reasons_col.loc[:, 18:21].max(axis=1)
type4 = reasons_col.loc[:, 22:].max(axis=1)
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
df = pd.concat([df, type1, type2, type3, type4], axis = 1)
df
```

```python
df.columns.values
```

```python
#It is more reasonable to change the column names from 1, 2, ... to Type 1, Type 2, ...
col_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
```

```python
df.columns = col_names
```

```python
df
```

```python
#To maintain original syntax and for convenience the reason columns need to be shifted to the beginning of the dataframe.

col_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[col_names_reordered]
```

```python
df.head(10)
```

```python
#It is advisable to create checkpoints since it is very useful in case past dataframe contents are required.
df_reasons = df.copy()
```

```python
#Now is a perfect time to convert the datatype of the date column
df_reasons['Date'] = pd.to_datetime(df_reasons['Date'], format='%d/%m/%Y')
df_reasons.info()
```

```python
#Months and days of the dates might be useful to check if there are any patterns, i.e., if employees are more 
#likely to be absent on specific months.
def dateToWeekday(dt):
    return dt.weekday()
def dateToMonth(dt):
    return dt.month
df_reasons['Month'] = df_reasons['Date'].apply(dateToMonth)
df_reasons['Day'] = df_reasons['Date'].apply(dateToWeekday)
```

```python
df_reasons
```

```python
df_date = df_reasons.copy()
```

```python
df_date = df_date.drop(['Date'], axis=1)
```

```python
df_date.columns.values
```

```python
col_list=['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month', 'Day',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_date=df_date[col_list]
```

```python
df_date
```

```python
#Other columns are self explanatory. Employees far away from workplace, high transportation expenses, daily work load, age
#high BMI and more number of children and pets are more likely to be absent. 
#Education again refers to categorical data. So it'll have to be changed to a dummy variable like in the case of 'Reason for
#Absence'.
```

```python
#Categories of Education column
#1: High School, 2: Graduate, 3: Postgraduate, 4: Master (PhD)
```

```python
df_date['Education'].value_counts()
```

```python
#Clearly most of the employees have only high school education. So it would be useless to categorize them into 4 types.
#Instead, we'll give 0 to every employee with high school education only and 1 to the rest.
df_date['Education'] = df_date['Education'].map({1:0, 2: 1, 3: 1, 4: 1})
```

```python
df_date['Education'].value_counts()
```
